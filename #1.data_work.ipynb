filling null values 

```python
df = pd.DataFrame(dataset)
from sklearn.impute import SimpleImputer

# List of categorical columns with nulls
cat_cols = [
    'Age_band_of_driver', 'Educational_level', 'Driving_experience', 
    'Type_of_vehicle', 'Road_surface_type' ,
    'Service_year_of_vehicle', 'Defect_of_vehicle', 'Area_accident_occured',
    'Lanes_or_Medians', 'Road_allignment', 'Types_of_Junction',
    'Type_of_collision', 'Vehicle_movement'
]

# Mode imputer for categorical columns
cat_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

# Numerical columns (if any nulls)
num_cols = ['Number_of_vehicles_involved']  # in your case no nulls, but safe to include
num_imputer = SimpleImputer(strategy='median')
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# Replace all 'Unknown' or 'unknown' with np.nan first
df.replace(['Unknown', 'unknown'], np.nan, inplace=True)
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
mode_imputer = SimpleImputer(strategy='most_frequent')
df[cat_cols] = mode_imputer.fit_transform(df[cat_cols])



print(df.isnull().sum())
df.isnull().sum().sum() #(0)
print("Total entries after imputation:", df.shape[0] * df.shape[1]) #246320
```
